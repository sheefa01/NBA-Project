```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("/tmp/nba_stats.sqlite")

tables = pd.read_sql("""
SELECT name 
FROM sqlite_master
WHERE type='table'
""", conn)

display(tables)
```

```python
for table in tables['name']:
    print("TABLE:", table)
    display(pd.read_sql(f"SELECT * FROM {table} LIMIT 5", conn))
```

```python
for table in tables['name']:
    count = pd.read_sql(f"SELECT COUNT(*) as rows FROM {table}", conn)
    print(table)
    display(count)
```

```python
for table in tables['name']:
    df = pd.read_sql(f"SELECT * FROM {table} LIMIT 1", conn)
    print("\n", table)
    print(df.columns.tolist())
```

```python
for table in tables['name']:
    df = pd.read_sql(f"SELECT * FROM {table}", conn)

    print("\nTABLE:", table)
    display(df.isnull().sum())
```

```python
for table in tables['name']:
    df = pd.read_sql(f"SELECT * FROM {table}", conn)

    print(table, "duplicates:", df.duplicated().sum())
    
```

```python
for table in tables['name']:
    df = pd.read_sql(f"SELECT * FROM {table}", conn)

    print("\nTABLE:", table)
    display(df.describe())
```
